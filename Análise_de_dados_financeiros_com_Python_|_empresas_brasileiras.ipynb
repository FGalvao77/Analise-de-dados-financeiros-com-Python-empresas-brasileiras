```python
import pandas as pd
import numpy as np
import yfinance as yf

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stocks = dict(
    PBR = 'Petróleo Brasileiro S.A. - Petrobras',
    ERJ = 'Embraer S.A.',
    BBD = 'Banco Bradesco S.A.',
    BSBR = 'Banco Santander (Brasil) S.A.',
    ITUB = 'Itaú Unibanco Holding S.A.',
    MGLUY = 'Magazine Luiza S.A.',
    EBR = 'Centrais Elétricas Brasileiras S.A. - Eletrobrás',
    VALE = 'Vale S.A.',
    USNZY = 'Usinas Siderúrgicas de Minas Gerais S.A.',
    BDORY = 'Banco do Brasil S.A.',
    LTMAQ = 'LATAM Airlines Group S.A.',
    AZUL = 'Azul S.A.',
    GOL = 'Gol Linhas Aéreas Inteligentes S.A.',
    GGB = 'Gerdau S.A.',
    SID = 'Companhia Siderúrgica Nacional',
    CIG = 'Companhia Energética de Minas Gerais'
)
```

```python
stocks.keys()
```

```python
df = yf.download(list(stocks))
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
adj_close = df['Adj Close']
```

```python
adj_close.head(20)
```

```python
adj_close.tail(20)
```

```python
adj_close.dropna().plot(figsize=(18,10));
```

```python
adj_close.dropna().plot(subplots=True, figsize=(18,12));
```

```python
adj_close = adj_close.dropna()
```

```python
for key, value in stocks.items():
    print(f'{key:6s} | {value}')
```

```python
adj_close.info()
```

```python
adj_close.describe().round(2).T
```

```python
adj_close.mean()
```

```python
adj_close.aggregate([min, np.mean, np.std, np.median, max]).round(2).T
```

### **Muda com o tempo**

```python
adj_close.diff(5).head(10)
```

```python
adj_close.diff().mean().head()
```

```python
adj_close.pct_change().round(3).head()
```

```python
adj_close.pct_change().mean().plot(kind='bar', figsize=(18,8));
```

```python
print(
    adj_close.head(),
    adj_close.shift(1).head(),
    sep='\n\n'
)
```

```python
rets = np.log(adj_close / adj_close.shift(1))
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(18,12));
```

```python
rets.cumsum().apply(np.exp).plot(subplots=True, figsize=(18,12));
```

### **Reamostragem**

```python
# 1 semana
adj_close.resample('1w', label='right').last().head().round(2)
```

```python
# 1 semana (média)
adj_close.resample('1w', label='right').mean().head().round(2)
```

```python
# 1 mês
adj_close.resample('1m', label='right').last().head().round(2)
```

```python
# 1 mês (média)
adj_close.resample('1m', label='right').mean().head().round(2)
```

```python
# 1 ano
adj_close.resample('1y', label='right').last().head().round(2)
```

```python
# 1 ano (média)
adj_close.resample('1y', label='right').mean().head().round(2)
```

```python
rets.cumsum().apply(np.exp).resample('1y', label='right').plot(figsize=(18,12));
```

```python
rets.cumsum().apply(np.exp).resample('1w', label='right').last().plot(figsize=(18,12));
```

```python
rets.cumsum().apply(np.exp).resample('1m', label='right').last().plot(figsize=(18,12));
```
